```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#merge the two datasets
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

```python
#to find the basic info of the data
movies.info()
```

```python
movies = movies[['movie_id','title','genres','overview','keywords','cast','crew',]]
```

```python
movies.head(1)
```

```python
movies.iloc[0].genres
```

```python
#to convert the dictionary to a list 
import ast
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies.head(1)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.cast[0]
```

```python
#we will use only the first three cast members
import ast
def convert3(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
import ast
def Director(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(Director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:str(x).split())
```

```python
movies['overview']
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies
```

```python
movies = movies[['movie_id', 'title', 'tags']]
```

```python
movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
#vectorizationv
vectors = cv.fit_transform(movies["tags"]).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():        #converting the data into a list
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#calculate the distance between each vector 
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
